```python
import numpy as np
import pandas as pd

from model import ThomasFieringGenerator

from load_data import load_example_data
from plotting import plot_flow_ranges, plot_fdc_ranges
```

```python
### Load demo data
test_site_idx = 0
Q_obs = load_example_data().iloc[:, test_site_idx].dropna()
Q_obs = Q_obs.loc[Q_obs > 0]
Q_obs_monthly = Q_obs.resample('MS').sum()

## ThomasFiering Model
tf = ThomasFieringGenerator(Q_obs_monthly)
Q_syn = tf.generate_ensemble(n_years=20, n_realizations=10)
Q_syn_tf = Q_syn.copy()

# Plot the results
plot_flow_ranges(Q_obs_monthly, Q_syn, timestep='monthly', savefig=False)
plot_fdc_ranges(Q_obs_monthly, Q_syn, savefig = False)
```
